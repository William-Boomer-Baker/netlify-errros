Removing 's' in animation delay and animation duration to get numerical value

```python
for x in df.index:
  df.loc[x, "animation-delay"] = df.loc[x, "animation-delay"].replace("s", "")
  df.loc[x, "animation-duration"] = df.loc[x, "animation-duration"].replace("s", "")
  # print(new_df.loc[x, "animation-delay"])
print(df.to_string()) 

```

Setting up background-color to numerical values

```python
def rgb_int (r,g,b):
  packed = int('%02x%02x%02x' % (r, g, b), 16)
  return packed

def rgba_int (r,g,b, a):
  packed = int('%02x%02x%02x%02x' % (r, g, b, a), 16)
  return packed


for x in df.index:
  # print(df.loc[x, "background-color"])
  # print(df.loc[x, "background-color"][df.loc[x, "background-color"].find('(')+1 : df.loc[x, "background-color"].find(')')])
  chr = (df.loc[x, "background-color"][df.loc[x, "background-color"].find('(')+1 : df.loc[x, "background-color"].find(')')]).split(', ')
  # print(len(chr), chr[0], chr[1], chr[2])
  if(len(chr) == 3):
    print("rgb", rgb_int(int(chr[0]), int(chr[1]), int(chr[2])))
    df.loc[x, "background-color"] = rgb_int(int(chr[0]), int(chr[1]), int(chr[2]))

  if(len(chr) == 4):
    print("rgba", rgba_int(int(chr[0]), int(chr[1]), int(chr[2]), int(chr[3])))
```
